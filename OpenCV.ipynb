```python
import cv2
import imutils
from matplotlib import pyplot as plt
```

# Bare minimum

```python
#load the image
image = cv2.imread("person-on-a-bridge-near-a-lake-747964.jpg")

#to open image in window
cv2.imshow("Image", image) 
#window remains open
cv2.waitKey(0)

#to display image in notebook
converted_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #v imp to convert BGR to RGB
# plt.xticks([])
# plt.yticks([])
plt.imshow(converted_img)
plt.show()

(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
```

```python
#colour input is in BGR but reflected as RGB since image has been converted to display in notebook
output=converted_img.copy()
cv2.rectangle(output, (1750, 1250), (2250,1750), (255,0,0), 5) #top-left, bottom-right,colour, width
cv2.circle(output, (300, 150), 100, (255, 0, 0), -1) #center, radius, colour, -ve to fill +ve for width
cv2.line(output, (3000, 2050), (4000, 3000), (255, 0, 0), 20) #similar to rectangle
plt.imshow(output)
plt.show()
```

```python
output = converted_img.copy()
cv2.putText(output, "Test Text", (1500, 250), cv2.FONT_HERSHEY_SIMPLEX, 10, (0,0,0), 10) #starting point, font, scale, colour, width
plt.imshow(output)
plt.show()
```

## Conversion to gray scale

```python
image = cv2.imread("jigsaw.png")
converted_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #v imp to convert BGR to RGB
# plt.xticks([])
# plt.yticks([])
plt.imshow(converted_img)
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray) 
cv2.waitKey(0)
```

## Edge Detection

```python
edged = cv2.Canny(gray, 30, 150) #minimum, maximum threshold
cv2.imshow("Edged", edged)
cv2.waitKey(0)
```
